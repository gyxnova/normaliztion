```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("wine_data.csv",usecols=[0,1,2])
df.head()
```

```python
sns.kdeplot(data=df,x="alcohol",hue="class_label")
plt.show()
```

```python
sns.kdeplot(data=df,x="malic_acid",hue="class_label")
plt.show()
```

```python
sns.scatterplot(x= "alcohol",y="malic_acid",data=df,hue = "class_label")
plt.show
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop("class_label",axis=1),
                                                 df["class_label"],test_size=0.3,
                                                 random_state=0)

x_train.shape,x_test.shape
```

```python
x_train_columns = x_train.columns
x_test_coloumns = x_test.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Convert the scaled NumPy arrays back to DataFrames using the original column names
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```
